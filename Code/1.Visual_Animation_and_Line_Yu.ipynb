# 2. Plot

### 2.1 Make the Grid

We will first define a list of string years which will represent the values that our slider will take on. Going through the dataset, we will take out all the unique values from the column `grade` and store them as well. Finally, we make a grid with each column representing a slice of the dataframe by `year`, `grade` and `column name`, making sure to name each column uniquly by these variables.

```python
# Make list of years
years = data.year.value_counts().keys().sort_values().astype(str).tolist()
years
```

```python
# Make list of grades
grades = data.grade.value_counts().keys().sort_values().tolist()
grades
```

```python
# Online mode!

columns = []
# make grid
for year in years:
    for grade in grades:
        data_by_year_and_grade = data[(data['year'] == int(year)) & (data['grade'] == grade)]
        for col_name in data_by_year_and_grade:
            # each column name is unique
            column_name = '{year}_{grade}_{header}_gapminder_grid'.format(
                year=year, grade=grade, header=col_name
            )
            a_column = Column(list(data_by_year_and_grade[col_name]), column_name)
            columns.append(a_column)
```

```python
grid = Grid(columns)
url = py.grid_ops.upload(grid, 'gapminder_grid_0425'+str(time.time()), auto_open=False)
url
```

### 2.2 Make the Figure

```python
figure = {
    'data': [],
    'layout': {},
    'frames': [],
    'config': {'scrollzoom': True}
}

# fill in most of layout
figure['layout']['xaxis'] = {'range':[25000,174500],'title': 'Average Annual Income', 'gridcolor': '#FFFFFF'}
figure['layout']['yaxis'] = {'range':[1500,28000],'title': 'Average Loan Amount','gridcolor': '#FFFFFF'}
figure['layout']['hovermode'] = 'closest'
figure['layout']['plot_bgcolor'] = 'rgb(223, 232, 243)'
```

### 2.3 Add Slider

For the slider to appear, we need to add a sliders dictionary to layout. The sliders dictionary is set in the following way:

```python
sliders_dict = {
    'active': 0,
    'yanchor': 'top',
    'xanchor': 'left',
    'currentvalue': {
        'font': {'size': 20},
        'prefix': 'Year:',
        'visible': True,
        'xanchor': 'right'
    },
    'transition': {'duration': 300, 'easing': 'cubic-in-out'},
    'pad': {'b': 10, 't': 50},
    'len': 0.9,
    'x': 0.1,
    'y': 0,
    'steps': []
}
```

### 2.4 Add Play and Pause Buttons

```python
figure['layout']['updatemenus'] = [
    {
        'buttons': [
            {
                'args': [None, {'frame': {'duration': 500, 'redraw': False},
                         'fromcurrent': True, 'transition': {'duration': 300, 'easing': 'quadratic-in-out'}}],
                'label': 'Play',
                'method': 'animate'
            },
            {
                'args': [[None], {'frame': {'duration': 0, 'redraw': False}, 'mode': 'immediate',
                'transition': {'duration': 0}}],
                'label': 'Pause',
                'method': 'animate'
            }
        ],
        'direction': 'left',
        'pad': {'r': 10, 't': 87},
        'showactive': False,
        'type': 'buttons',
        'x': 0.1,
        'xanchor': 'right',
        'y': 0,
        'yanchor': 'top'
    }
]

custom_colors = {
    'A': 'rgb(171, 99, 250)',
    'B': 'rgb(230, 99, 250)',
    'C': 'rgb(99, 110, 250)',
    'D': 'rgb(25, 211, 243)',
    'E': 'rgb(50, 170, 255)',
    'F': 'rgb(210, 211, 243)',
    'G': 'rgb(50, 190, 115)'
}
```

### 2.5 Fill in Figure with Data and Frames

```python
col_name_template = '{year}_{grade}_{header}_gapminder_grid'
```

```python
year = 2007
for grade in grades:
    data_dict = {
        'xsrc': grid.get_column_reference(col_name_template.format(
            year=year, grade=grade, header='avgIncome'
        )),
        'ysrc': grid.get_column_reference(col_name_template.format(
            year=year, grade=grade, header='avgAmt'
        )),
        'mode': 'markers',
        'textsrc': grid.get_column_reference(col_name_template.format(
            year=year, grade=grade, header='sub_grade'
        )),
        'marker': {
            'sizemode': 'area',
            'sizeref': 0.1,
            'sizesrc': grid.get_column_reference(col_name_template.format(
                 year=year, grade=grade, header='avgIntRate'
            )),
            'color': custom_colors[grade]
        },
        'name': grade
    }
    figure['data'].append(data_dict)
```

### 2.6 Create Frames

Finally we make our frames. Here we are running again through the years and grades, but for each combination we instantiate a frame dictionary of the form: <br>
`frame = {'data': [], 'name': value-name}`

We add a dictionary of data to this list and at the end of each loop, we ensure to add the steps dictionary to the steps list. At the end, we attatch the sliders dictionary to the figure via: <br>
`figure['layout']['sliders'] = [sliders_dict]`

```python
for year in years:
    frame = {'data': [], 'name': str(year)}
    for grade in grades:
        data_dict = {
            'xsrc': grid.get_column_reference(col_name_template.format(
                year=year, grade=grade, header='avgIncome'
            )),
            'ysrc': grid.get_column_reference(col_name_template.format(
                year=year, grade=grade, header='avgAmt'
            )),
            'mode': 'markers',
            'textsrc': grid.get_column_reference(col_name_template.format(
                year=year, grade=grade, header='sub_grade'
            )),
            'marker': {
                'sizemode': 'area',
                'sizeref': 0.1,
                'sizesrc': grid.get_column_reference(col_name_template.format(
                 year=year, grade=grade, header='avgIntRate'
                )),
                'color': custom_colors[grade]
            },
            'name': grade
        }
        frame['data'].append(data_dict)

        
    figure['frames'].append(frame)
    slider_step = {'args': [
        [year],
        {'frame': {'duration': 300, 'redraw': False},
         'mode': 'immediate',
       'transition': {'duration': 300}}
     ],
     'label': year,
     'method': 'animate'}
    sliders_dict['steps'].append(slider_step)

figure['layout']['sliders'] = [sliders_dict]
```

### 2.7 Plot Animation

```python
py.icreate_animations(figure, 'gapminder_example'+str(time.time()))
```

# 1. Plot

```python
df=data2_final.copy().drop(['num_loans','count'],axis=1)
df=df.pivot_table(values='percent',index=['year'],columns='grade').fillna(0)
df.to_csv('subplots_pivot_table.csv')
df=pd.read_csv('subplots_pivot_table.csv',header=0)
```

```python
df
```

```python
# Make list of years
years = df.year.value_counts().keys().sort_values().astype(str).tolist()
years
```

```python
# Make list of purposes
grades = data2_final.grade.value_counts().keys().sort_values().tolist()
grades
```

```python
title = "Main Grade for Loans"
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
```

```python
colors = ['rgb(171, 99, 250)', 
          'rgb(230, 99, 250)', 
          'rgb(99, 110, 250)', 
          'rgb(25, 211, 243)',
          'rgb(50, 170, 255)',
            'rgb(210, 211, 243)',
            'rgb(50, 190, 115)'
         ]
```

```python
traces = []

for i,grade in enumerate(grades):
    traces.append(go.Scatter(
        x=years,
        y=df[grade].values.tolist(),
        name = grade,
        mode='lines',
        line=dict(color=colors[i], width=2),
        connectgaps=True,
    ))

    traces.append(go.Scatter(
        x=['2007', '2015'],
        y=[df[grade].values.tolist()[0], 
           df[grade].values.tolist()[-1]],
        name=grade,
        mode='markers',
        marker=dict(color=colors[i], size=9)
    ))
```

```python
layout = go.Layout(
    xaxis=dict(
        showline=True,
        showgrid=False,
        showticklabels=True,
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        autotick=False,
        ticks='outside',
        tickcolor='rgb(204, 204, 204)',
        tickwidth=2,
        ticklen=5,
        tickfont=dict(
            family='Arial',
            size=16,
            color='rgb(82, 82, 82)',
        ),
    ),
    yaxis=dict(
        showgrid=False,
        zeroline=False,
        showline=False,
        showticklabels=False,
    ),
    autosize=False,
    margin=dict(
        autoexpand=False,
        l=100,
        r=20,
        t=110,
    ),
    showlegend=False,
)
```

```python
annotations = []

# Adding labels
for y_trace, label, color in zip(np.array(df.T[1:]).round().astype(int).tolist(), labels, colors):
    # labeling the left_side of the plot
    annotations.append(dict(xref='paper', x=0.05, y=y_trace[0],
                                  xanchor='right', yanchor='middle',
                                  text=label + ' {}%'.format(y_trace[0]),
                                  font=dict(family='Arial',
                                            size=15,
                                            color=colors),
                                  showarrow=False))
    # labeling the right_side of the plot
    annotations.append(dict(xref='paper', x=0.95, y=y_trace[-1],
                                  xanchor='left', yanchor='middle',
                                  text='{}%'.format(y_trace[-1]),
                                  font=dict(family='Arial',
                                            size=15,
                                            color=colors),
                                  showarrow=False))
# Purposes
annotations.append(dict(xref='paper', yref='paper', x=0.5, y=-0.1,
                              xanchor='center', yanchor='top',
                              text='Lending Club Loan Data',
                              font=dict(family='Arial',
                                        size=12,
                                        color='rgb(150,150,150)'),
                              showarrow=False))

layout['annotations'] = annotations
```

```python
fig = go.Figure(data=traces, layout=layout)
py.iplot(fig, filename='news-source')
```
